```python
import pandas as pd
import numpy as np
```

```python
PATH = '..\data\cnes_swarm50\Traces50.csv'

df = pd.read_csv(PATH, header=None)
df
```

```python
df['sat_id'] = [e for sat in [[i]*3 for i in range(50)] for e in sat]
df
```

```python
for i in range(50):
    dfi = df[df['sat_id']==i]
    dfi.to_csv('..\data\cnes_swarm50\\track_'+str(i)+'.csv', index=False, columns=np.arange(0,10000,1))
```
