```python
def plot_groups(swarms, title=None, show_edges=True):
    fig = plt.figure(figsize=(10,10))
    ax = plt.axes(projection='3d')
    colordict = {
        0:'tab:blue',
        1:'tab:orange',
        2:'tab:green',
        3:'tab:red',
        4:'tab:purple',
        5:'tab:brown',
        6:'tab:pink',
        7:'tab:gray',
        8:'tab:olive',
        9:'tab:cyan',
        -1:'black'
    }
    
    for gid, swarm in swarms.items():
        x_data = [node.x for node in swarm.nodes]
        y_data = [node.y for node in swarm.nodes]
        z_data = [node.z for node in swarm.nodes]
        sc = ax.scatter(x_data, y_data, z_data, 
                        c=colordict[gid], 
                        s=45, 
                        edgecolors='black',
                        label=str(gid))
        if show_edges:
            for node in swarm.nodes:
                for n in node.neighbors:
                    if n.group==gid:
                        ax.plot([node.x, n.x], [node.y, n.y], [node.z, n.z], 
                                c=colordict[gid]
                                )
    
    ax.legend(title='Groups')
    
    ax.set_title(title)
```

```python
def diameter(swarm, group, default=-1):
    G = swarm.swarm_to_nxgraph()
    node_ids = [n.id for n in group.nodes]
    max_length = (0,0,0) # Source, target, number of hops
    for ni in node_ids:
        for nj in node_ids:
            nb_hops = default
            if nx.has_path(G, ni, nj):
                nb_hops = len(nx.shortest_path(G, ni, nj)) -1
            if nb_hops > max_length[2]:
                max_length = (ni, nj, nb_hops)
    return max_length
```

```python
def top_n_sources_reversed(swarm, n=10):
    p = {} #proba of each node to become a source
    degrees = swarm.degree()
    for node in swarm.nodes:
        ndeg = degrees[node.id]
        p[node.id] = 0.
        if ndeg > 0:
            p[node.id] = 1./ndeg
    return dict(sorted(p.items(), key=lambda x:x[1], reverse=True)[:n])
```

```python
def param_MDRW(swarm, src_ids, s=1, overlap=False):
    sources = [swarm.get_node_by_id(id) for id in src_ids]
    swarms = {} # Dict(group ID:Swarm)
    for i,src in enumerate(sources): # Initialize swarms
        src.set_group(i)
        swarms[i] = Swarm(swarm.connection_range, nodes=[src])
    free_nodes = [n for n in swarm.nodes if n.group==-1]
    while free_nodes: # Spread paths to desired length
        for k in swarms.keys():
            n_i = swarms[k].nodes[-1] # Current node
            free_neighbors = set(free_nodes).intersection(n_i.neighbors)
            if free_neighbors: # At least one unassigned neighbor
                n_j = n_i.random_walk(s, overlap) # Next node
            else:
                n_j = swarm.random_jump(s, overlap) # If no neighbor, perform random jump in the graph
            n_j.set_group(n_i.group)
            swarms[k].add_node(n_j) 
            free_nodes.remove(n_j)
    return swarms
    
```

```python
def top_n_sources(swarm, n=10):
    degrees = {}
    for node in swarm.nodes:
        degrees[node.id] = node.degree()
    return dict(sorted(degrees.items(), key=lambda x:x[1], reverse=True)[:n])
```

```python
# Agony optimization : https://arxiv.org/pdf/1902.01477.pdf 

def agony(n1, n2):
    d1, d2 = n1.degree(), n2.degree()
    if d2 >= d1:
        return -100
    else: 
        return d2 + 1 - d1

def get_scores(n_i, nodes):
    score = {}
    for n_j in nodes:
        score[n_j] = agony(n_i, n_j)
    return score

def MDRW_A(swarm, src_ids, s=1, overlap=False):
    sources = [swarm.get_node_by_id(id) for id in src_ids]
    swarms = {} # Dict(group ID:Swarm)
    for i,src in enumerate(sources): # Initialize swarms
        src.set_group(i)
        swarms[i] = Swarm(swarm.connection_range, nodes=[src])
    free_nodes = [n for n in swarm.nodes if n.group==-1]
    while free_nodes: # Some nodes do not belong to a group yet
        for k in swarms.keys():
            n_i = swarms[k].nodes[-1] # Current node
            free_neighbors = set(free_nodes).intersection(n_i.neighbors)
            if free_neighbors: # At least one unassigned neighbor
                scores = get_scores(n_i, free_neighbors)
                best_nodes = [n for n in free_neighbors if scores[n] == np.max(list(scores.values()))]
                next_node = choice(best_nodes) # Next node
            else:
                next_node = swarm.random_jump(s, overlap) # If no neighbor, perform random jump in the graph
            next_node.set_group(n_i.group)
            swarms[k].add_node(next_node) 
            free_nodes.remove(next_node)
    return swarms
```

```python
# Test on a single group
swarm = swarm_case[1505]
top10 = top_n_sources(swarm, n=10)
src_ids = list(top10.keys())
swarm.reset_groups()
print('Top 10 degree sources:', src_ids)
print('What\'s happening for group 0 from source', src_ids[0], '?\n')
    
src = swarm.get_node_by_id(src_ids[0])
src.set_group(0)
print('Source degree:', src.degree())
swarm_0 = Swarm(swarm.connection_range, nodes=[src])

free_nodes = [n for n in swarm.nodes if n.group==-1]
for i in range(10): # Pick 10 nodes for the group
    n_i = swarm_0.nodes[-1] # Current node
    free_neighbors = set(free_nodes).intersection(n_i.neighbors)
    if free_neighbors: # At least one unassigned neighbor
        print('List of free neighbors:', [n.id for n in free_neighbors])
        scores = get_scores(n_i, free_neighbors)
        print('Neighbor agony scores:')
        for node, score in scores.items():
            print(node.id, 'degree', node.degree(), 'scoring', score)
        best_nodes = [n for n in free_neighbors if scores[n] == np.max(list(scores.values()))]
        print('Best nodes:', [n.id for n in best_nodes])
        next_node = choice(best_nodes) # Next node
        print('Choosing', next_node.id,'\n')
    else:
        next_node = swarm.random_jump(s=1, overlap=False) # If no neighbor, perform random jump in the graph
        print('Random jump. Choosing', next_node.id)
    next_node.set_group(n_i.group)
    swarm_0.add_node(next_node) 
    free_nodes.remove(next_node)
print('Done!')
print(swarm_0)
for n in swarm_0.nodes:
    print(n.id, n.degree())
```
